```python
#訓練データを読み込む
df_train = pd.read_csv(path + "data/train.csv")
pd.options.display.max_columns = None
df_train.head()
```

```python
X_orig = df_train.iloc[:, 1:-1].drop('matchType', axis=1)
y_orig = df_train.iloc[:, -1]
display(X_orig.head())
display(y_orig.head())
```

```python
X_train = df_train.drop(["winPlacePerc","Id","matchType"],axis=1)
display(X_train.head())
```

```python
#学習モデルを構築する（各自で工夫する）
#df_train = df_train.loc[:, ["rideDistance", "swimDistance", "walkDistance", "winPlacePerc"]]
X_train = df_train.drop(["winPlacePerc","Id","matchType"],axis=1)
#display(df_train.head())
#X_train = df_train.drop("winPlacePerc",axis=1)
display(X_train.head())
#X_train = df_train.iloc[:, :-1].values
y_train = df_train.winPlacePerc
#y_train = df_train.iloc[:, -1].values
X_orig_train, X_orig_test, y_orig_train, y_orig_test = train_test_split(X_train, y_train, train_size=0.7, random_state=00)
lr = LinearRegression()
lr.fit(X_orig_train, y_orig_train)
print('Train score: {}'.format(mean_absolute_error(y_orig_train, lr.predict(X_orig_train))))
print('Test score: {}'.format(mean_absolute_error(y_orig_test, lr.predict(X_orig_test))))
rfr_orig = RandomForestRegressor(random_state=42)
rfr_orig.fit(X_orig_train, y_orig_train)
print('Train score: {}'.format(mean_absolute_error(y_orig_train, rfr_orig.predict(X_orig_train))))
print('Test score: {}'.format(mean_absolute_error(y_orig_test, rfr_orig.predict(X_orig_test))))
```

```python
#学習モデルを構築する（各自で工夫する）
#df_train = df_train.loc[:, ["rideDistance", "swimDistance", "walkDistance", "winPlacePerc"]]
X_train = pd.concat([df_train.drop(["winPlacePerc","Id","matchType"],axis=1),pd.get_dummies(df_train.matchType,drop_first=True)],axis=1)
#display(df_train.head())
#X_train = df_train.drop("winPlacePerc",axis=1)
display(X_train.head())
#X_train = df_train.iloc[:, :-1].values
y_train = df_train.winPlacePerc
#y_train = df_train.iloc[:, -1].values
X_orig_train, X_orig_test, y_orig_train, y_orig_test = train_test_split(X_train, y_train, train_size=0.7, random_state=42)
lr = LinearRegression()
lr.fit(X_orig_train, y_orig_train)
print('Train score: {}'.format(mean_absolute_error(y_orig_train, lr.predict(X_orig_train))))
print('Test score: {}'.format(mean_absolute_error(y_orig_test, lr.predict(X_orig_test))))
rfr_orig = RandomForestRegressor(random_state=42)
rfr_orig.fit(X_orig_train, y_orig_train)
print('Train score: {}'.format(mean_absolute_error(y_orig_train, rfr_orig.predict(X_orig_train))))
print('Test score: {}'.format(mean_absolute_error(y_orig_test, rfr_orig.predict(X_orig_test))))
```

```python
rfr_ohe = RandomForestRegressor(max_depth=None, min_samples_leaf=4, random_state=42)
rfr_ohe.fit(X_orig_train, y_orig_train)
print("Train score: {}".format(mean_absolute_error(y_orig_train, rfr_ohe.predict(X_orig_train))))
print("Test score: {}".format(mean_absolute_error(y_orig_test, rfr_ohe.predict(X_orig_test))))
```

```python
#グリッドサーチによるパラメータの探索
#時間がかなりかかります
paramJ1_1 = {'depth': list(range(9, 10, 1)),
             'learning_rate': list(np.arange(0.04, 0.06, 0.01))}
#paramJ1_1 = {'n_estimators': list(range(800, 1001, 100)),
#             'learning_rate': list(np.arange(0.10, 0.11, 0.1))}

mae_scoring = make_scorer(mean_absolute_error, greater_is_better=False)

cbr_orig = CatBoostRegressor(random_state=42)

X_orig_train_fdel = X_orig_train.drop(X_orig_train.columns[[16,19,23,26,27,28,30,31,32]],axis=1)
X_orig_test_fdel = X_orig_test.drop(X_orig_train.columns[[16,19,23,26,27,28,30,31,32]],axis=1)

#rfr_orig_gs = GridSearchCV(RandomForestRegressor(random_state=42), rfr_orig_params, scoring=mae_scoring, cv=4)
#gsearch1 = GridSearchCV(estimator = mod1J1,param_grid = paramJ1_1,cv = 5,n_jobs=4,scoring = 'neg_mean_squared_error')
cbr_orig_gs = GridSearchCV(estimator = cbr_orig,param_grid = paramJ1_1,cv = 5,n_jobs=4,scoring=mae_scoring)
cbr_orig_gs.fit(X_orig_train_fdel, y_orig_train)
print("Train score: {}".format(-cbr_orig_gs.score(X_orig_train_fdel, y_orig_train)))
print("Test score: {}".format(-cbr_orig_gs.score(X_orig_test_fdel, y_orig_test)))
print("Best parameters: {}".format(cbr_orig_gs.best_params_))
print("Best cross-validation: {}".format(-cbr_orig_gs.best_score_))
```
